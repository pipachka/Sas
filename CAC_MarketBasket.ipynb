Считаем какие категории товаров покупаются чаще всего вместе 

```python
import itertools

info_mb = ['support', 'confidence', 'lift']
resulting_info_mb = pd.DataFrame(columns=info_mb)

perms = list(' & '.join(i) for i in itertools.permutations(columns_mb, 2))
```

```python
len(perms)
```

```python
resulting_info_mb.insert(0, 'Пара категорий', perms)
resulting_info_mb.fillna(0, inplace=True)
resulting_info_mb.head(10)
```

Сделали пустую таблицу, теперь заполняем

```python
def support(perm, data):
    counter = 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(data.shape[0]):
        if data[fir][i] > 0 and data[sec][i] > 0:
            counter += 1
    return counter / len(data)

def confidence(perm, data):
    counter_both, counter_first = 0, 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(data.shape[0]):
        if data[fir][i] > 0: 
            counter_first += 1
            if data[sec][i] > 0:
                counter_both += 1
    return counter_both/counter_first

def denom(perm, data):
    counter_first = 0
    counter_second = 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(data.shape[0]):
        if data[fir][i] > 0:
            counter_first += 1
        if data[sec][i] > 0:
            counter_second += 1
    return (counter_first * counter_second) / data.shape[0]**2
```

```python
# resulting_info_mb.loc[resulting_info_mb['Пара категорий'] == perms[i], 'support'] = sup
```

```python
for i in range(len(perms)):
    sup = support(perms[i], market_basket) 
    resulting_info_mb.loc[resulting_info_mb['Пара категорий'] == perms[i], 'support'] = sup
    
    conf = confidence(perms[i], market_basket) 
    resulting_info_mb.loc[resulting_info_mb['Пара категорий'] == perms[i], 'confidence'] = conf
    
    lift = sup / denom(perms[i], market_basket)
    resulting_info_mb.loc[resulting_info_mb['Пара категорий'] == perms[i], 'lift'] = lift
resulting_info_mb.head(10)
```

Отсортируем, ведь support у A->B и B->A одинаковый

## MarketBasket по кластерам
номер кластера смотрим в labels = 0,1,2,3

```python
import itertools
```

```python
data = pd.read_csv('vitrina_with_clusters_4.csv', sep=';') 
data.head(10) 
```

```python
data[data['ИнтернетДоставка'] < 1]
```

```python
df_1 = data[data['Labels'] == 0]
df_2 = data[data['Labels'] == 1]
df_3 = data[data['Labels'] == 2]
df_4 = data[data['Labels'] == 3]

df_1 = df_1.reset_index(drop=True)
df_2 = df_2.reset_index(drop=True)
df_3 = df_3.reset_index(drop=True)
df_4 = df_4.reset_index(drop=True)

display(df_1[df_1['Кол-во уникальных клиентов_buyed'] == 1])
df_1.columns
```

Кластер 1

```python
info_mb = ['support', 'confidence', 'lift']
resulting_info_mb_1 = pd.DataFrame(columns=info_mb)

resulting_info_mb_1.insert(0, 'Пара категорий', perms)
resulting_info_mb_1.fillna(0, inplace=True)

perms = list(' & '.join(i) for i in itertools.permutations(columns_mb, 2))
```

```python
def support(perm, data):
    counter = 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(len(data)):
        if data.loc[i, fir] > 0 and data.loc[i, sec] > 0:
            counter += 1
    return counter / len(data)

def confidence(perm, data):
    counter_both, counter_first = 0, 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(data.shape[0]):
        if data[fir][i] > 0: 
            counter_first += 1
            if data[sec][i] > 0:
                counter_both += 1
    if counter_first == 0:
        return 0
    return counter_both/counter_first

def denom(perm, data):
    counter_first = 0
    counter_second = 0
    fir =  perm.split(' & ')[0]
    sec = perm.split(' & ')[1]
    for i in range(data.shape[0]):
        if data[fir][i] > 0:
            counter_first += 1
        if data[sec][i] > 0:
            counter_second += 1
    return (counter_first * counter_second) / data.shape[0]**2
```

```python
for i in range(len(perms)):
    sup = support(perms[i], df_1) 
    resulting_info_mb_1.loc[resulting_info_mb_1['Пара категорий'] == perms[i], 'support'] = sup
    
    conf = confidence(perms[i], df_1) 
    resulting_info_mb_1.loc[resulting_info_mb_1['Пара категорий'] == perms[i], 'confidence'] = conf
    
    if denom(perms[i], df_1) != 0:
        lift = sup / denom(perms[i], df_1)
    else:
        lift = 0
    resulting_info_mb_1.loc[resulting_info_mb_1['Пара категорий'] == perms[i], 'lift'] = lift
resulting_info_mb_1.head(10)
```

```python
resulting_info_mb_1.sort_values(by=['support', 'confidence'], ascending=False).head(10)
```

```python
resulting_info_mb_1.to_csv('MarketBasket_cluster_1.csv', index=False, sep=';', encoding='utf-8-sig') 
```
